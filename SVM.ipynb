## **Implementing SVM**

```python
from sklearn.model_selection import train_test_split
```

### **Split the data**
For the independent data we will use al the features in df_feat and for the y data we will use the target.

```python
X = df_feat
y = cancer['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

### **Instance and train the model**

```python
from sklearn.svm import SVC
model = SVC(gamma='auto')
```

```python
model.fit(X_train, y_train)
```

```python
model.get_params()
```

### **Make predictions**

```python
predictions = model.predict(X_test)
```

### **Evaluate the model**

In this first evaluation, we can see the accuracy and results are quite bad. But there is a tool for selecting the best parameters for the c, gamma and other parameters.

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test, predictions))
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, predictions).ravel()
df_confusion = pd.DataFrame(
    [[tn, fp], [fn, tp]],
    columns = ["Predicted Negative", "Predicted Positive"],
    index = ["Actual Negative", "Actual Positive"]
)
df_confusion
```

```python
accuracyScore = accuracy_score(y_test, predictions)
error_rate = 1-accuracyScore

print('Accuracy: ', np.round(accuracyScore*100,2))
print('Error rate: ', np.round(error_rate*100,2))
```

## **Grid Search CV**

---

This is a tool that helps finding the best possible settings for a machine learning model. We usually just guess the values of the settings and change them based on the outcome, but Grid Search CV will get the best settings.

For it to work we have to provide it with various possible settings (in this case C, Gamma and Kernel) and will run it various times and test the best settings.

```python
from sklearn.model_selection import GridSearchCV
```

First we need to pass the list of parameters we want to test.

```python
param_grid = {'C': [0.1,1,10,100,1000], 'gamma' : [1,0.1,0.01,0.001]}
```

```python
grid = GridSearchCV(SVC(), param_grid, verbose=2)
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

After analysing all of that data, Grid Search CV has established that the best value for our SVM model is C = 1 and gamma = 0.001. Let's run another evaluation of the grid predictions and see what the scores are.

```python
grid_predictions = grid.predict(X_test)
```

```python
print(classification_report(y_test, grid_predictions))
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, grid_predictions).ravel()
df_confusion = pd.DataFrame(
    [[tn, fp], [fn, tp]],
    columns = ["Predicted Negative", "Predicted Positive"],
    index = ["Actual Negative", "Actual Positive"]
)
df_confusion
```

```python
accuracyScore = accuracy_score(y_test, grid_predictions)
error_rate = 1-accuracyScore

print('Accuracy: ', np.round(accuracyScore*100,2))
print('Error rate: ', np.round(error_rate*100,2))
```

We can see that the parameters chosen by our Grid Search CV are significantly better than what we guessed the first time.
