### *Read salaries.csv as a data frame called **sal** and answer the following:*


1.   What is the name of the highest paid person (including benefits)?
2.   What was the avg (mean) BasePay of all employees per year (2011-2014)?
3.   How many unique job titles are there?
4.   What are the top 5 most common jobs?
5.   How many job titles where represented by only one person in 2013?



Import libraries

```python
import numpy as np
import pandas as pd
```

Read the CSV file

```python
sal = pd.read_csv('/content/salaries.csv')
sal.head()
```

## **1-**

The highest paid person (total pay benefits): Nathaniel Ford

```python
sal.sort_values('TotalPayBenefits', ascending=False).head(1)
```

# **2-**

```python
sal.pivot_table(index='Year', values='BasePay', aggfunc='mean')
```

## **3-**

```python
print('{0} unique jobs'.format(sal['JobTitle'].nunique()))
```

## **4-**

```python
sal['JobTitle'].value_counts().head()
```

```python
selectYear = sal[sal['Year'] == 2013]
jobCount = selectYear['JobTitle'].value_counts()

onePerson = len(jobCount[jobCount == 1])

print(f'Number of job titles with only one person in 2013: {onePerson}')
```
