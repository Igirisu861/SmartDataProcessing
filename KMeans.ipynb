# **K-means Algorithm**

---

This is an unsupervised machine learning algorithm which clusters different unlabeled data around centroids.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## **Generating artificial data**

---
### **Blobs**
* A blob is a group of points which are close to each other.
* Each blob represents a cluster of similar data points
* We can control how many blobs you want, how they spread out and where they are located

```python
import sklearn.datasets as make_blobs
```

### **Parameters:**
* Number of samples
* Number of features
* How many centers
* The standard deviation of the cluster
* The random seed


```python
data = make_blobs.make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
type(data)
```

To get the samples

```python
data[0]
```

To see the labels (number of clusters)

```python
data[1]
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='Pastel2')
```

## **Implementing K-Means**

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4, random_state=101)
```

```python
kmeans.fit(data[0])
```

With this, we receive the value of each one of the centers

```python
kmeans.cluster_centers_
```

And the labels...

```python
kmeans.labels_
```

### **Comparing the K-means clustering against the original data labels**

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,6))
ax1.set_title('K means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='Pastel2')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='Pastel2')
```

```python
fig, ax1 = plt.subplots(figsize=(10,6))
ax1.set_title('K means')
ax1.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='Pastel2')

ax1.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='black', marker='*', edgecolors='black', label = "Centroids")
ax1.legend()
```

### **Comparing with K = 3**

```python
kmeans3 = KMeans(n_clusters=3, random_state=101)
kmeans3.fit(data[0])
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,6))
ax1.set_title('K means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans3.labels_, cmap='Pastel2')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='Pastel2')
```

## **SSE (Sum of Squared Error)**

```python
sse = []

for i in range(1, 10):
  km = KMeans(n_clusters=i, random_state=101)
  km.fit(data[0])
  sse.append(km.inertia_)
  print("For n_clusters = {} SSE = {}".format(i, km.inertia_))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,10), sse, color='purple',linestyle='--', marker='*', markerfacecolor='lavender', markersize = 10)
plt.title('SSE vs K Value')
plt.xlabel('K')
plt.ylabel('SSE')
```
